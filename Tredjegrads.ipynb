```python
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt
```

```python
def demand(Q):
    return 29000-40*Q
```

```python
def marginalrevenue(Q):
    return 29000 - (80*Q)
```

```python
def marginalcost(price):
    return 5000 +(2*Q)
```

```python
Q=sp.symbols('Q', real=True, positive=True)
equ=sp.Eq(marginalrevenue(Q),marginalcost(Q))
equ
```

Når vi setter MR = MC kan vi finne det optimale kvantumet slik

```python
Q_equ=sp.solve(equ,Q)[0]
round(Q_equ, 2)
```

```python
p_eq=demand(Q_equ)
round(p_eq,2)
```

```python
#drawing 800 points in the interval 0.0 to 800
Q = np.linspace(1,1000,1000)

#drawing marginalrevenue
plt.plot(Q,marginalrevenue(Q),label='MR')

#drawing marginalcost
plt.plot(Q,marginalcost(Q),label='MC')
#drawing demand
plt.plot(Q,demand(Q),color='green',label='Etterspørsel')
#
plt.plot([0, 292.68], [17292.68, 17292.68], color='k', linestyle='dashed', linewidth=2, label='Norsk marked')
plt.plot([292.68, 292.68], [0, 17292.68], color='k', linestyle='dashed', linewidth=2)
#adding legend:
plt.legend(loc='upper right')

#setting the axis lables
plt.ylabel('Pris')
plt.xlabel('Passasjerer')

#setting the range of y values (the price), so that we can more
#easily see the equilibrium price
plt.ylim((0, 30000))
plt.xlim((0, 1000))
```

```python
def Tyskdemand(Q):
    return 47000-52*Q
```
