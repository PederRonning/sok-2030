```python
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt
```

```python
def demand(Q):
    return (29000-40*Q)
```

```python
def marginalrevenue(Q):
    return 29000 - (80*Q)
```

```python
def marginalcost(price):
    return 5000 +(2*Q)
```

```python
Q=sp.symbols('Q', real=True, positive=True)
equ=sp.Eq(marginalrevenue(Q),marginalcost(Q))
equ
```

Når vi setter MR = MC kan vi finne det optimale kvantumet slik

```python
Q_equ=sp.solve(equ,Q)[0]
round(Q_equ, 2)
```

```python
Q = np.linspace(1,1000,1000)

#drawing marginalrevenue
plt.plot(Q,marginalrevenue(Q),label='MR')

#drawing marginalcost
plt.plot(Q,marginalcost(Q),label='MC')
#drawing demand
plt.plot(Q,demand(Q),color='green',label='Etterspørsel')
#
plt.plot([0, 292.68], [17468.8, 17468.8], color='k', linestyle='dashed', linewidth=2, label='Norsk marked')
plt.plot([292.68, 292.68], [0, 17468.8], color='k', linestyle='dashed', linewidth=2)
#adding legend:
plt.legend(loc='upper right')

plt.ylabel('Pris')
plt.xlabel('Passasjerer')

plt.ylim((0, 30000))
plt.xlim((0, 1000))
```

```python
#Priselastisitet
P1 = 17468.8
Delta_P = 10 # endring i pris på 10%
P2 = P1 * (1 + Delta_P/100)

Q1 = demand(P1)
Q2 = demand(P2)

priselastisitetnorge = ((Q2 - Q1) / ((Q1 + Q2) / 2)) / (Delta_P/100)
```

```python
round(priselastisitetnorge, 3)
```

```python
def Tyskdemand(Q):
    return 47000-52*Q
```

```python
def Tyskmarginalrevenue(Q):
    return 47000 - (104*Q)
```

```python
Q=sp.symbols('Q', real=True, positive=True)
Tyskequ=sp.Eq(Tyskmarginalrevenue(Q),marginalcost(Q))
Tyskequ
```
