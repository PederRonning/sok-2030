```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def demand_tyskland(p):
    return ((47000 - p)/52)

def demand_norge(p):
    return ((29000 - p)/40)

def m_demand(p):
    w=demand_tyskland(p)
    z=demand_norge(p)
    sum = (z<0)*w + (z>=0)*(z+w)

        
    return sum
```

```python
# Tegne individuell etterspørsel og den horisontale summen

p_num = np.linspace(0,50000,100)

fig, axs = plt.subplots(1,3, figsize=(20,5))

# plott av linjene

axs[1].plot(demand_norge(p_num),p_num,label='Norsk etterspørsel')
axs[0].plot(demand_tyskland(p_num),p_num,label='Tysk etterspørsel')
axs[2].plot(m_demand(p_num),p_num,label='Aggregert Markedsetterspørsel')

# bruk løkke for å tegne aksene

for i in range(3):

    axs[i].set_ylabel('Pris (kr)')
    axs[i].set_xlabel('Passasjerer')
    axs[i].set(ylim=(0,50000),xlim=(0,2000))
    axs[i].spines['top'].set_color('none')
    axs[i].spines['right'].set_color('none')
    axs[i].legend(loc='best', frameon=False)

# gi en passende tittel

fig.suptitle('Utledning av Hurtigrutens etterspørsel i markedet')

plt.show()
```
