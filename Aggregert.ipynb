```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
def demand_tyskland(p):
    return ((47000 - p)/52)

def demand_norge(p):
    return ((29000 - p)/40)

def m_demand(p):
    w=demand_tyskland(p)
    z=demand_norge(p)
    sum = (z<0)*w + (z>=0)*(z+w)

        
    return sum
```

```python
# Tegne individuell etterspørsel og den horisontale summen

p_num = np.linspace(0,50000,100)

fig, axs = plt.subplots(1,3, figsize=(20,5))

# plott av linjene

axs[1].plot(demand_norge(p_num),p_num,label='Norsk etterspørsel')
axs[0].plot(demand_tyskland(p_num),p_num,label='Tysk etterspørsel')
axs[2].plot(m_demand(p_num),p_num,label='Aggregert Markedsetterspørsel')

# bruk løkke for å tegne aksene

for i in range(3):

    axs[i].set_ylabel('Pris (kr)')
    axs[i].set_xlabel('Passasjerer')
    axs[i].set(ylim=(0,50000),xlim=(0,2000))
    axs[i].spines['top'].set_color('none')
    axs[i].spines['right'].set_color('none')
    axs[i].legend(loc='best', frameon=False)

# gi en passende tittel

fig.suptitle('Utledning av Hurtigrutens etterspørsel i markedet')

plt.show()
```

```python
def demandhigh(Q):
    return 45997.6-52*Q
```

```python
def mrhigh(Q):
    return 46997.6 - (104*Q)
```

```python
def mrlow(Q):
    return 37018.2 - (45.46*Q)
```

```python
def marginalcost(price):
    return 5000 +(2*Q)
```

## MR=MC

```python
Q=sp.symbols('Q', real=True, positive=True)
equ1=sp.Eq(mrhigh(Q),marginalcost(Q))
equ1
```

```python
Q=sp.symbols('Q', real=True, positive=True)
equ2=sp.Eq(mrlow(Q),marginalcost(Q))
equ2
```

### Optimal tilpasning når 47 000 ≥ P ≥ 29 000

```python
#Kvantum/passasjerer
Q_equ1=sp.solve(equ1,Q)[0]
round(Q_equ1, 2)
```

```python
#Pris
p_eq1=demandhigh(Q_equ1)
round(p_eq1,2)
```

### Optimal tilpasning når P < 29 000

```python
#Kvantum/passasjerer
Q_equ2=sp.solve(equ2,Q)[0]
round(Q_equ2, 2)
```

```python
#Pris
p_eq2=demandhigh(Q_equ2)
round(p_eq2,2)
```

```python
def demand2(p):
    return 903.8-(p/52)
```

```python
def demand3(p):
    return 1626.8-(0.044*p)
```

```python
def marginalrevenuehigh(p):
    return (46997.6 - p) / 104
```

```python
def marginalrevenuelow(p):
    return (37018.2 - p) / 45.46
```

```python
def MC(p):
    return (p - 5000) / 2
```
